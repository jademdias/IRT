```python
clf1o = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf1o.csv')
clf1o.drop(clf1o.tail(1).index,inplace=True)
clf2o = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf2o.csv')
clf3o = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf3o.csv')
clf4o = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf4o.csv')
clf5o = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf5o.csv')
clf6o = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf6o.csv')


clf1a = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf1a.csv')
clf2a = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf2a.csv')
clf3a = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf3a.csv')
clf4a = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf4a.csv')
clf5a = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf5a.csv')
clf6a = pd.read_csv('/content/drive/Shareddrives/TRI/Grafos/clf6a.csv')
```

```python
dataori = ['Original']*100
names = ["Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]
clf1o.loc[:,'Data'] = dataori
clf1o.loc[:,'classifier'] = names[0]
clf2o.loc[:,'Data'] = dataori
clf2o.loc[:,'classifier'] = names[1]
clf3o.loc[:,'Data'] = dataori
clf3o.loc[:,'classifier'] = names[2]
clf4o.loc[:,'Data'] = dataori
clf4o.loc[:,'classifier'] = names[3]
clf5o.loc[:,'Data'] = dataori
clf5o.loc[:,'classifier'] = names[4]
clf6o.loc[:,'Data'] = dataori
clf6o.loc[:,'classifier'] = names[5]

dataaug = ['Augmented']*100
clf1a.loc[:,'Data'] = dataaug
clf2a.loc[:,'Data'] = dataaug
clf3a.loc[:,'Data'] = dataaug
clf4a.loc[:,'Data'] = dataaug
clf5a.loc[:,'Data'] = dataaug
clf6a.loc[:,'Data'] = dataaug
```

```python
m = pd.merge(clf1a, clf2a, how = 'outer')
m = pd.merge(m, clf3a, how = 'outer')
m = pd.merge(m, clf4a, how = 'outer')
m = pd.merge(m, clf5a, how = 'outer')
m = pd.merge(m, clf6a, how = 'outer')
m = pd.merge(m, clf1o, how = 'outer')
m = pd.merge(m, clf2o, how = 'outer')
m = pd.merge(m, clf3o, how = 'outer')
m = pd.merge(m, clf4o, how = 'outer')
m = pd.merge(m, clf5o, how = 'outer')
m = pd.merge(m, clf6o, how = 'outer')
m
```
